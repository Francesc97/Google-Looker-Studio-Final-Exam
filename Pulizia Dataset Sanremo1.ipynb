```python
import pandas as pd
import re

def parse_periodo_semplice(periodo_str):
    """
    Estrae data inizio e fine basandosi su:
    - DATA INIZIO: primo numero + prima parola di sole lettere
    - DATA FINE: secondo numero + prima parola di sole lettere dopo il secondo numero
    """
    if pd.isna(periodo_str) or periodo_str == '':
        return None, None
    
    periodo = str(periodo_str).strip()
    
    # Trova tutti i numeri nella stringa
    numeri = re.findall(r'\d+', periodo)
    
    # Trova tutte le parole composte solo da lettere
    parole = re.findall(r'[a-zA-Zàáâäçéèêëíìîïñóòôöúùûüýÿ]+', periodo)
    
    if len(numeri) < 2 or len(parole) < 1:
        return periodo, periodo
    
    # DATA INIZIO: primo numero + prima parola
    data_inizio = f"{numeri[0]} {parole[0]}"
    
    # DATA FINE: secondo numero + prima parola 
    data_fine = f"{numeri[1]} {parole[0]}"
    
    # Se ci sono più parole, la seconda potrebbe essere per la data fine
    if len(parole) > 1:
        data_fine = f"{numeri[1]} {parole[1]}"
    
    return data_inizio, data_fine
```

```python
import pandas as pd
import re

def parse_periodo_semplice(Periodo_str):
    if pd.isna(Periodo_str) or Periodo_str == '':
        return None, None
    
    Periodo = str(Periodo_str).strip()
    
    # Trova tutti i numeri nella stringa
    numeri = re.findall(r'\d+', Periodo)
    
    # Trova tutte le parole composte solo da lettere
    parole = re.findall(r'[a-zA-Zàáâäçéèêëíìîïñóòôöúùûüýÿ]+', Periodo)
    
    if len(numeri) < 2 or len(parole) < 1:
        return Periodo, Periodo
    
    # DATA INIZIO: primo numero + prima parola
    data_inizio = f"{numeri[0]} {parole[0]}"
    
    # DATA FINE: secondo numero + prima parola 
    data_fine = f"{numeri[1]} {parole[0]}"
    
    # Se ci sono più parole, la seconda potrebbe essere per la data fine
    if len(parole) > 1:
        data_fine = f"{numeri[1]} {parole[1]}"
    
    return data_inizio, data_fine

# Applicazione sul DataFrame df

date_parsed = df['Periodo'].apply(parse_periodo_semplice)
df['data_inizio'] = date_parsed.apply(lambda x: x[0])
df['data_fine'] = date_parsed.apply(lambda x: x[1])
```

```python
def sostituisci_multiplo(mese, numeri):
    """
    Sostituisce più stringhe in un testo utilizzando un dizionario.

    Args:
        testo: La stringa in cui effettuare le sostituzioni.
        sostituzioni: Un dizionario dove le chiavi sono le stringhe da sostituire
                      e i valori sono le stringhe con cui sostituirle.

    Returns:
        La stringa risultante con tutte le sostituzioni applicate.
    """
    for vecchia_stringa, nuova_stringa in sostituzioni.items():
        mese = mese.replace(mese, numero)
    return testo
```
